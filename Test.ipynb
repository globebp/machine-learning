# **Demo ML project**

## This multi-variate analysis demo project will predict logS value based on the other four columns using Linear Regression and Randon forest techniques

### Load Data




```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/dataprofessor/data/master/delaney_solubility_with_descriptors.csv")
df
```

### Seperation x columns (inputs) and y(target)


```python
y = df['logS']
y
```

```python
x = df.drop('logS',  axis=1)
x
```

### Split the Data Set for training and testing (20%) to be able to test and train on different data sets 

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=100)
```

```python
X_train
```

```python
X_test
```

```python
y_train
```

```python
y_test
```

### Build the model


## **Linear regression model**


### Train the model

```python

from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train, y_train)

LinearRegression()

```

### Apply the model


```python
y_lr_train_pred = lr.predict(X_train)
y_lr_test_pred = lr.predict(X_test)
```

### y prediction from training data set

```python
y_lr_train_pred
```

### y prediction from test data set

```python
y_lr_test_pred
```

### Evaluate Model performance by comparing the mean squared error between the predicted values and the actual values of a regression


```python
from sklearn.metrics import mean_squared_error, r2_score

lr_train_mse = mean_squared_error(y_train, y_lr_train_pred)
lr_train_r2 = r2_score(y_train, y_lr_train_pred)

lr_test_mse = mean_squared_error(y_test, y_lr_test_pred)
lr_test_r2 = r2_score(y_test, y_lr_test_pred)


```

```python
lr_results = pd.DataFrame(['Liner Regression', lr_train_mse, lr_train_r2, lr_test_mse, lr_test_r2]).transpose()

lr_results.columns = ['Method', 'Training MSE', 'Training R2', 'Test MSE', 'Test R2']

lr_results

```

## **Random Forest**


## Train the model


```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor (max_depth=2, random_state=100)
rf.fit(X_train, y_train)


```

## Apply the model to make predictions


```python

y_rf_train_pred = rf.predict(X_train)
y_rf_test_pred = rf.predict(X_test)

```

### Evaluate Model performance

```python
from sklearn.metrics import mean_squared_error, r2_score

rf_train_mse = mean_squared_error(y_train, y_rf_train_pred)
rf_train_r2 = r2_score(y_train, y_rf_train_pred)

rf_test_mse = mean_squared_error(y_test, y_rf_test_pred)
rf_test_r2 = r2_score(y_test, y_rf_test_pred)
```

```python
rf_results = pd.DataFrame(['Random Forest', rf_train_mse, rf_train_r2, rf_test_mse, rf_test_r2]).transpose()

rf_results.columns = ['Method', 'Training MSE', 'Training R2', 'Test MSE', 'Test R2']

rf_results
```

### Display the performance of Linear regression and random forest in one table for comparision


```python
df_models = pd.concat([lr_results, rf_results], axis=0)

df_models.reset_index()
```

## Data visualization of the predicted results

```python
import matplotlib.pyplot as plt
import numpy as np

plt.scatter(x=y_train, y=y_lr_train_pred, alpha=0.2)

z= np.polyfit(y_train, y_lr_train_pred, 1)
p = np.poly1d(z)

plt.plot(y_train, p(y_train), 'red')

plt.ylabel('Predict logS')
plt.xlabel('Experimental logS')
```
